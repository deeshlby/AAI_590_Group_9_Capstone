```python
import pandas as pd
import glob
import os

# Define the path to the directory containing the CSV files
folder_path = r"C:\Users\eterp\OneDrive\Documents\USD\AAI_590_Capstone\HCL150"  # Replace with the correct path to your folder
```

```python
# Get a list of all CSV files in the directory
csv_files = glob.glob(os.path.join(folder_path, "*.csv"))

# Dictionary to hold DataFrames for each CSV file
dataframes = {}

# Load each CSV file into a DataFrame and store it in the dictionary
for file in csv_files:
    file_name = os.path.basename(file)  # Get the name of the file
    df = pd.read_csv(file)  # Read the CSV file into a DataFrame
    dataframes[file_name] = df  # Store the DataFrame in the dictionary with the file name as the key
```

```python
# Preview the first few rows of each DataFrame
for file_name, df in dataframes.items():
    print(f"\n--- Preview for {file_name} ---")
    print(df.head())
```
